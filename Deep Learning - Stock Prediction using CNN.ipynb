create a function to do the labeling

```python
import numpy as np
def labelling(data):
  window_size = 11
  counter_row = 0
  number_of_days = len(data)
  list_label = []
  while counter_row<number_of_days:
    counter_row = counter_row+1
    if counter_row > window_size:
      window_begin_index = counter_row - window_size
      window_end_index = window_begin_index + window_size - 1
      window_middle_index = (window_begin_index+window_end_index)/2
      min = 100000
      max = 0
      for i in range (window_begin_index, window_end_index+1):
        number = data['Close'][i]
        if number < min:
          min = number
          min_index = i
        if number > max:
          max = number
          max_index = i
      if max_index == window_middle_index:
        label = 'Sell'
      elif min_index == window_middle_index:
        label = 'Buy'
      else:
        label = 'Hold'
      list_label.append(label)
    else :
      list_label.append(np.nan)
  return list_label
```

### Arsitektur CNN

```python
import tensorflow as tf
from tensorflow.keras.datasets import cifar10
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, Activation, Flatten
from tensorflow.keras.layers import Conv2D, MaxPooling2D

# import pickle

# pickle_in = open("X.pickle","rb")
# X = pickle.load(pickle_in)

# pickle_in = open("y.pickle","rb")
# y = pickle.load(pickle_in)

# X = X/255.0

n1 = datetime.now()

model = Sequential()

model.add(Conv2D(64, (3, 3), input_shape=X_train.shape[1:], padding='same'))
# model.add(Conv2D(64, (2, 2), input_shape=X_train.shape[1:]))
model.add(Activation('relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))

model.add(Conv2D(64, (3, 3), padding='same'))
# model.add(Conv2D(64, (2, 2)))
model.add(Activation('relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))

model.add(Flatten())  # this converts our 3D feature maps to 1D feature vectors

model.add(Dense(64, activation = 'relu'))
model.add(Dense(64, activation = 'relu'))
model.add(Dense(3, activation = 'softmax'))
# model.add(Activation('softmax'))

model.compile(loss='categorical_crossentropy',
              optimizer='SGD',
              metrics=['accuracy'])



# history = model.fit(X, Y, validation_split=0.33, epochs=150, batch_size=10, verbose=0)

n2 = datetime.now()
print('waktu:',n2-n1)
```

### Evaluasi Model

Prediksi

```python
prediction = model.predict(X_test)
pred = np.argmax(prediction, axis=1)
# prediction
```

Akurasi

```python
y_test_index = np.argmax(y_test, axis = 1)
benar = 0
for i in range (len(pred)):
  if pred[i]==y_test_index[i]:
    benar+=1
akurasi = benar/len(pred)
print('Akurasi:',akurasi)
```

Confusion Matrix

```python
from sklearn.metrics import confusion_matrix

#Generate the confusion matrix
cf_matrix = confusion_matrix(y_test_index, pred)
# cf_matrix = confusion_matrix(pred , y_test_index)

print(cf_matrix)
```

```python
import seaborn as sns

ax = sns.heatmap(cf_matrix, annot=True, cmap='Blues')

ax.set_title('Seaborn Confusion Matrix with labels\n\n');
ax.set_xlabel('\nPredicted Values')
ax.set_ylabel('Actual Values ');

## Ticket labels - List must be in alphabetical order
ax.xaxis.set_ticklabels(['Buy','Hold','Sell'])
ax.yaxis.set_ticklabels(['Buy','Hold','Sell'])

## Display the visualization of the Confusion Matrix.
plt.show()
```

Recall, Precision, F1 Score

```python
col_label = ['Hold','Buy','Sell']
recall_value = []
print('Recall:')
for i in range (len(col_label)):
    recall = cf_matrix[i][i]/cf_matrix[i].sum()
    print(col_label[i]+': '+str(recall))
    recall_value.append(recall)
precision_value = []
print('Precision:')
for i in range (len(col_label)):
    precision = cf_matrix.T[i][i]/cf_matrix.T[i].sum()
    print(col_label[i]+': '+str(precision))
    precision_value.append(precision)
f1_score_value = []
print('F1 Score:')
for i in range (len(col_label)):
    f1 = (2 * recall_value[i] * precision_value[i])/ ( recall_value[i] + precision_value[i])
    print(col_label[i]+': '+str(f1))
    f1_score_value.append(f1)


```

```python
evaluasi = pd.DataFrame({'Action':['Hold','Buy','Sell'],'Recall':recall_value,'Precision':precision_value,'F1 Score':f1_score_value})
evaluasi.set_index('Action')
```
